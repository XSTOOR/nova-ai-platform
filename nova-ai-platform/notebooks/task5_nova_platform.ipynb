# Task 5: Multi-Agent Platform — NOVA AI Support System

## Overview
LangGraph-based multi-agent orchestration integrating all previous tasks:

### Agent Architecture
```
Customer Message → Triage → Support/Recommendation → Escalation Check → Response
                     │              │                      │
                     │              │                      │
               Task 1:         Task 2+3+4            Task 1+2
               - Intent        - MCP Tools           - Escalation Logic
               - Injection      - RAG Pipeline        - Ticket Creation
               - Routing       - Brand Voice
```

### 6 Demo Scenarios
1. Order Status Inquiry → MCP order_lookup tool
2. Product Recommendation → RAG pipeline retrieval
3. Return Request → MCP return_initiate tool
4. Loyalty Points Check → MCP loyalty_check tool
5. Injection Attack Defense → Blocked by injection defender
6. Escalation to Human → Ticket creation with SLA

````
# ═══════════════════════════════════════════════════════════════
# Cell 10: Audit Trail Deep Dive
# ═══════════════════════════════════════════════════════════════
import json

print('🔍 Audit Trail Deep Dive')
print('='*60)

# Run a query and examine the full audit trail
result = graph.run(
    message='Where is my order ORD-2024-001?',
    customer_id='sarah@example.com',
    customer_name='Sarah',
)

print(f'Total audit steps: {len(result["audit_trail"])}')
print()

for i, entry in enumerate(result['audit_trail']):
    print(f'Step {i+1}: [{entry["agent"]}] {entry.get("action", "?")}')
    print(f'  Duration: {entry.get("duration_ms", 0):.1f}ms')
    details = {k: v for k, v in entry.items() 
               if k not in ('agent', 'step', 'action', 'timestamp', 'duration_ms')}
    if details:
        print(f'  Details: {json.dumps(details, indent=4, default=str)[:200]}')
    print()
````
